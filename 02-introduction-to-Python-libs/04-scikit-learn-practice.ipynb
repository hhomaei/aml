## Data Loading & Feature Selection

### Data Loading

````python
from sklearn import datasets

# Loading dataset
iris = datasets.load_iris() # return class:~sklearn.utils.Bunch (Dictionary-like object)
print("The dataset shape: ", iris.data.shape)
print("Feature names: ", iris.feature_names)
print("The iris species: ", iris.target_names)
````

````python
# Choose two features => The algorithm will decide about classification based on these features
x=iris.data[:, [2, 3]] # In this example we chose the 3rd and 4th features (columns)
y=iris.target
````

## Configuring the ML Algorithm


Assume that we want to use KNN (K Nearest Neighbors) algorithm to classify the iris flowers. We use scikit-learn KNN classifier and configure its parameters.<br>
KNN parameters:
- n_neighbors = number of neighbors to be considered for detection
- metric = How to calculate the distance between points
> - Most common distance meytrics:
>> 1. Euclidean distance: $\sqrt{\Sigma(x_i - y_i)^2}$
>> 2. Manhattan distance: $\Sigma|x_i - y_i|$
>> 3. Minkowski distance (general form of distance): $\sqrt[p]{\Sigma|x_i - y_i|^p}$
>>> - $p = 1 \rightarrow Manhattan$
>>> - $p = 2 \rightarrow Euclidean$
>>> - $p = \infty \rightarrow Chebyshev$ ($\lim _{p \rightarrow \infty}\left(\sum_{i=1}^n\left|x_i-y_i\right|^p\right)^{\frac{1}{p}}=\max _{i=1}^n\left|x_i-y_i\right|$). It means that we can go as a king of chess! In other words, all kinds of movements (horizontal, vertical, or diagonal) are one-steps.


````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)
````

## Training
Fit data to the model in order to train the model

````python
knn.fit(x, y)
````

## Prediction & Evaluation

````python
# Predicting the class of a new instance
import numpy as np
predicted_y = knn.predict([[5, 2], [2, 1], [3, 3]])
print("predicted labels for [5, 2] , [2, 1], and [3, 3] are", predicted_y)
````

````python
# Evaluation: testing the result

# spliting train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

# Train
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)
knn.fit(x_train, y_train)

# Test
knn.score(x_test, y_test) # Return the mean accuracy on the given test data and labels. Accuracy is a metric that measures how often a machine learning model correctly predicts the outcome.
````

train_test_split(), by default, split the whole data into the train and test partitions randomly. In other words, it suffles the data and hence we get different results in each run. We can use the argument suffle to configure this behavior.

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=False)
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
````

Why we get such a bad result? Because, the function with shuffle=False splits the data by simply slicing it. This means the training set is taken from the beginning of the dataset and the test set from the end. In the iris dataset, the data is sorted based on the species. So, It is entirely reasonable to arrive at that conclusion. See the data again:

````python
iris_df
````

Considering the above mentioned problem, we should select random samples. However, we prefer to get the same results each time running the code. To satisfy these two requirements simultanously, we can use the random_state argument in the train_test_split function. If we set the random_state parameter to a fix integer, the function randomly selects train and test data, but all runs work with same splits.

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state = 34)
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
````

A popular value chosen all over the world for the random state is 42. The number simply has been made popular by a comic science fiction called "The Hitchhiker's Guide to the Galaxy" in 1978.

In this novel, a supercomputer called takes seven and a half million years to compute the answer to the ultimate question of Life, The Universe and Everything. The answer turns out to be 42. But why we do not use 42 it the above example? Simply, because of the bad result!

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state = 42)
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
````

There is another argument, stratify, which makes sure that the proportion of samples does not change. e.g. if we have 100 labels including 40 L1 and 60 L2, 40% of selected samples are chosen from L1 and 60% from L2. We will discuss this option in the later lectures. The usage is ```stratify = axis (e.g. y)```

Now, let me ask another important question: Why n_neighbors set to 6?

The answer: Test it!

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state = 34)

# Test for different number of neighbors [1-30]:
test_range = 30
neighbors=np.arange(1,30)
test_accuracy=np.empty(test_range)
for k in range(1, test_range + 1):
    knn_model = KNeighborsClassifier(n_neighbors = k)
    knn_model.fit(x_train, y_train)
    test_accuracy[k-1] = knn_model.score(x_test, y_test)   
plt.plot(range(1, test_range + 1), test_accuracy, label='Test Accuracy')
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
````

The best accuracy is achieved for n_neighbors = 4.

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, random_state = 34)
knn = KNeighborsClassifier(n_neighbors = 4, metric = 'minkowski', p=2)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
````

It is perfect. Isn't it?

No. This leads the model to overfit the new samples. 

There are two important concepts is machine learning models:
- Under-fitting: too simple to explain the variance
- Over-fitting: too-good to be true (noise may be enumerated)

We will discuss these concepts in later lectures.

In the previous evaluations, we split the data into the test and training partitions. We saw that changing these partitions (e.g. by changing the random_state or even discarding it), will change the score. So, which score should we report?

The answer: Repeat the test and report the mean score. This method is called cross-validation.

k-fold cross-validation means splitting the (training) data into k folds (in this case, three), then training the model k times, holding out a different fold each time for evaluation.

Cross-validation is also used for hyperparameter selection. We will discuss this topic in later lectures.

````python
# Evaluation by k-folds cross-validation method
knn = KNeighborsClassifier(n_neighbors = 6, metric = 'minkowski', p=2)

from sklearn.model_selection import cross_val_score

# The cross_val_score() function repeats the train-test process K times
scores = cross_val_score(knn, x, y, cv=5) # CV = Number of folds (K) = 5
scores
````

````python
# Error = mean of errors for all iterations
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
````

## More

Lots of machine learning algorithms are implemented in the scikit-learn library. The application of these algorithms is just the same. Here, we describe two other algorithms to show this similarity: A supervised regression model, Linear Regression, and an unsupervised learning algorithm, K-means.

### Linear Regression

The linear regression algorithm tries to find the line that has, on average, the minimum distance to each observation. The line is said to be the "best-fitting line." Assume that the x and y show the feature and target respectively. The algorithm finds the best 'a' and 'b' in the linear equation $y = ax + b$. The best 'a' and 'b' should minimize the sum of the vertical distance between each point and the prediction line which is called Ordinary Least Squares (OLS).

More general case: For n different features, the algorithm should find n + 1 coefficients in $y = a_1x_1 + a_2x_2 + ... + a_nx_n + b$

````python
# Sample data
x = np.arange (1, 11)
y = np.array([17, 16, 18, 20, 21, 23, 23, 25, 24, 26])

# Convert vector (one-row) to one-column matrix
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Visualize data
plt.scatter(x,y)
plt.ylim(bottom=0)
````

````python
# Train the model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
# Predict for already exist Xs
y_predicted = reg.predict(x)
````

````python
# Visualize the result
plt.plot(x, y_predicted)
plt.scatter(x, y, c='red')
plt.ylim(bottom=0)
````

How good is our model? The lower the distance (error) between the predicted and real values, the better the model. So, we should calculate this error. RMSE (Root Mean Square Error) is a common error calculation method. RSME measures the average difference between the predicted values by the model and the actual values using the following formula.
$$\sqrt{\frac{1}{m} \sum_{i=1}^m(y^i - \hat y^i)^2}$$

````python
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y, y_predicted)
````

### K-means
K-Means clustering is an unsupervised ML algorithm, which groups the unlabeled dataset into different clusters based on their similarity. The term 'K' is a number. You need to tell the system how many clusters you need to create. For example, K = 3 refers to three clusters. K-means algorithm works as follows:
1. Initializing the set of cluster centroids: Randomly select k samples from data. The algorithm assumes that these points are the central points (centroids) of the clusters.
2. Assigning Observations to Clusters: Compute the distance (using Euclidean distance measure for instance) between centroids and all other samples and assign each observation to the centroid closest to it.
3. Updating the Centroids: Once all of the observations have been assigned to a cluster, we shift the centroid of each cluster to the mean of its assigned observations. 

Steps 2 and 3 are repeated for either a set number of iterations or until convergence, which occurs when the cluster centroids no longer change.

See the following figure \[[reference](https://www.linkedin.com/pulse/k-means-clustering-its-real-use-case-surayya-shaikh "K-Means Algorithm")\]:

<center><img src="pictures/k-means.gif" width="50%" height="50%"/></center>


Example: clustering the unlabeled iris flowers.

````python
# Load data
x=iris.data[:, [2, 3]]

# Specify the algorithm and parameters
from sklearn.cluster import KMeans
kmn = KMeans(n_clusters = 3, random_state = 42) #random_state determines random number generation for centroid initialization
kmn.fit(x)
````

````python
labels = kmn.predict(x)
labels
````

We can also combine fit and predict functions and use ```kmn.fit_predict(x)```, instead of kmn.fit(x) and kmn.predict(x) separately.

````python
kmn.fit_predict(x)
````

````python
# You can predict the cluster of new instances easily by comparing them with the centroids. 
# The cluster of the closest centroid to the new data will be the result of prediction.
X_new = np.array([[1, 0.75], [4, 1.5], [6.5, 2]])
predicted = kmn.predict(X_new)
predicted
````

````python
# We can also visualize the clusters and centroids
plt.scatter(x[:,0], x[:,1], c=labels)
centers = kmn.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], marker='x', s=150)
````

````python
# need a better visualization? Rename the labels and use them in the legend

# Assign new names to the labels
label_names = {0: "Cluster 0", 1: "Cluster 1", 2: "Cluster 2"}
[print(label_names[label]) for label in labels]

# Scatter plot with color-coded labels
for label in np.unique(labels):
    plt.scatter(x[labels == label, 0], x[labels == label, 1], label = label_names[label])

# Scatter plot for cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=150, c='black')

# Add a legend with new labels
plt.legend()
````

````python
# Which label is assigned to each sample?
kmn.labels_
````

````python
# Create separate dataframes with the samples from each cluster

# Create a dataframe from the features and labels
df = pd.DataFrame(x, columns=['petal_length', 'petal_width'])
df['label'] = labels

# Create separate dataframes for each cluster
df_cluster_0 = df[df['label'] == 0].drop(columns=['label'])
df_cluster_1 = df[df['label'] == 1].drop(columns=['label'])
df_cluster_2 = df[df['label'] == 2].drop(columns=['label'])

print("Samples with label 0:")
print(df_cluster_0.head())

print("Samples with label 1:")
print(df_cluster_1.head())

print("Samples with label 2:")
print(df_cluster_2.head())
````

Question: How can we find k (number of clusters)?<br>
Answer: Test different values in a loop and take the best one.

Question: Which one is the best? We need a metric to evaluate the best.<br>
Answer: Use Inertia. It is the sum of the squared distances between the instances and their closest centroids. A good model is one with low inertia AND a low number of clusters (more clusters means more complex model). However, this is a tradeoff because as K increases, inertia decreases.

````python
y=[]
for k in np.arange(1,6):
    kmn=KMeans(n_clusters=k)
    kmn.fit(x)
    y.append(kmn.inertia_)
print(y)

plt.plot(np.arange(1,6), y, 'o-')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
````

As you can see in the plot, 5 is the best because it is the lowest. However, We choose 3. Because the differences between 3, 4, and 5 are not large enough that is worth adding complexity to the model. Moreover, we already know that there exist only three species of iris flowers!!!
