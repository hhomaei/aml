# Gradient Descent

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## Example 1
In this example, we will work with the simple function:
$$f(x) = x^2$$
We will show how Gradient Descent works and why we cannot choose a large learning rate.

```python
def ex1_function(x):
    return x**2

def ex1_derivative(x):
    return 2 * x 

x_ex1 = np.arange(-50, 50, 0.1)
y_ex1 = ex1_function(x_ex1)

plt.plot(x_ex1, y_ex1);
```

```python
# Choosing a random initial point
random.seed(4321)
initial_x_ex1 = random.uniform(-100, 100)
initial_point_ex1 = (initial_x_ex1, ex1_function(initial_x_ex1))
initial_point_ex1
```

```python
plt.plot(x_ex1, y_ex1)
plt.scatter(initial_point_ex1[0], initial_point_ex1[1], c='red');
```

```python
import imageio.v2 as imageio
import os

def gradient_descent(example_number, func, derv, x, y, initial_point, learning_rate = 0.01, precision = 0.001, max_iteration = 200, frame_speed = 20):
    print("The Gradient Descent Algorithm is now running...")
    print(f"Learning rate = {learning_rate}")

    current_x = initial_point[0]
    current_y = initial_point[1]

    frames = []  # list to store frames

    for iteration in range(max_iteration):
        print(f"Iteration {iteration+1}: {current_x, current_y}")

        plt.plot(x, y)
        plt.scatter(current_x, current_y, c='red')
        plt.pause(1/frame_speed)

        # Capture the current figure as an image
        plt.savefig("temp.png")
        frames.append(imageio.imread("temp.png"))

        plt.clf()

        # Save the last f(x)
        previous_y = current_y

        current_x = current_x - learning_rate * derv(current_x)
        current_y = func(current_x)
        if (np.isnan(current_y) or np.isneginf(current_y) or np.isinf(current_y)):
            print("The point was thrown into an out-of-range environment!")
            break

        if abs(current_y - previous_y) <= precision:
            break

    # Save the last frame
    plt.plot(x, y)
    plt.scatter(current_x, current_y, c='red')
    plt.pause(1/frame_speed)
    plt.savefig("temp.png")
    frames.append(imageio.imread("temp.png"))
    plt.clf()

    # Save all frames as a GIF
    imageName = f"gradient_descent_ex{example_number}_{learning_rate}"
    imageio.mimsave(f"images/{imageName}.gif", frames, fps = frame_speed, loop = 0)
    
    print(f"The final point: {current_x, current_y}")

    # Delete temporal objects
    frames.clear()
    os.remove("temp.png")
    # Close the interactive figure environment
    plt.close("all")
```

```python
# To show an interactive figure, I switch to an interactive backend of matplotlib (TkAgg)
%matplotlib tk

gradient_descent(1, ex1_function, ex1_derivative, x_ex1, y_ex1, initial_point_ex1, learning_rate = 0.01, max_iteration = 300)
```

```python
gradient_descent(1, ex1_function, ex1_derivative, x_ex1, y_ex1, initial_point_ex1, learning_rate = 0.1)
```

```python
gradient_descent(1, ex1_function, ex1_derivative, x_ex1, y_ex1, initial_point_ex1, learning_rate = 0.9)
```

```python
gradient_descent(1, ex1_function, ex1_derivative, x_ex1, y_ex1, initial_point_ex1, learning_rate = 1)
```

```python
gradient_descent(1, ex1_function, ex1_derivative, x_ex1, y_ex1, initial_point_ex1, learning_rate = 1.1, max_iteration = 50)
```

## Example 2
Let's see a complex function:
$$f(x) = e^x-ln(x)$$
The derivative of function $f(x)$ is:
$$f'(x) = e^x-1/x$$

```python
def ex2_function(x):
    return np.exp(x) - np.log(x)

def ex2_derivative(x):
    return np.exp(x) - 1/x 
```

```python
x_ex2 = np.arange(0.0000001, 3, 0.01) # logarithm is defined only for x > 0
y_ex2 = ex2_function(x_ex2)

plt.plot(x_ex2, y_ex2);
```

# Example 3
In this example we will rty to show that the initial point is important.
The following function has multiple minima:
$$f(x)=x\sin(x)$$
The derivative:
$$f'(x)=\sin(x)+x\cos(x)$$

```python
def ex3_function(x):
    return x * np.sin(x)

def ex3_derivative(x):
    return np.sin(x) + x*np.cos(x)

x_ex3 = np.arange(-20, 20, 0.1)
y_ex3 = ex3_function(x_ex3)

initial_point_ex3_1 = (-1.5, ex3_function(-1.8))
initial_point_ex3_2 = (14, ex3_function(14))

plt.plot(x_ex3, y_ex3)
plt.scatter(initial_point_ex3_1[0], initial_point_ex3_1[1], c='red');
plt.scatter(initial_point_ex3_2[0], initial_point_ex3_2[1], c='red');
```

```python
%matplotlib tk

gradient_descent(31, ex3_function, ex3_derivative, x_ex3, y_ex3, initial_point_ex3_1)
```

```python
gradient_descent(32, ex3_function, ex3_derivative, x_ex3, y_ex3, initial_point_ex3_2)
```

# Newton's Method

## Example 1
Function:
$$f(x) = x^2$$
The derivative of function $f(x)$ is:
$$f'(x) = 2x$$
The second derivative of function $f(x)$ is:
$$f''(x) = 2$$

```python
def ex1_derivative_2(x):
    return 2
```

```python
def newton_method(example_number, func, derv, derv_2, x, y, initial_point, precision = 0.001, max_iteration = 200, frame_speed = 20):
    print("The # Newton's Method is now running...")

    current_x = initial_point[0]
    current_y = initial_point[1]

    frames = []  # list to store frames

    for iteration in range(max_iteration):
        print(f"Iteration {iteration+1}: {current_x, current_y}")

        plt.plot(x, y)
        plt.scatter(current_x, current_y, c='red')
        plt.pause(1/frame_speed)

        # Capture the current figure as an image
        plt.savefig("temp.png")
        frames.append(imageio.imread("temp.png"))

        plt.clf()

        # Save the last f(x)
        previous_y = current_y

        current_x = current_x - derv(current_x)/derv_2(current_x)
        current_y = func(current_x)

        if abs(current_y - previous_y) <= precision:
            break

    # Save the last frame
    plt.plot(x, y)
    plt.scatter(current_x, current_y, c='red')
    plt.pause(1/frame_speed)
    plt.savefig("temp.png")
    frames.append(imageio.imread("temp.png"))
    plt.clf()

    # Save all frames as a GIF
    imageName = f"newton_method_ex{example_number}"
    imageio.mimsave(f"images/{imageName}.gif", frames, fps = frame_speed, loop = 0)
    
    print(f"The final point: {current_x, current_y}")

    # Delete temporal objects
    frames.clear()
    os.remove("temp.png")
    # Close the interactive figure environment
    plt.close("all")
```

```python
%matplotlib tk

newton_method(1, ex1_function, ex1_derivative, ex1_derivative_2, x_ex1, y_ex1, initial_point_ex1, frame_speed = 1)
```

## Example 2
The function:
$$f(x) = e^x-ln(x)$$
The derivative of function $f(x)$ is:
$$f'(x) = e^x-1/x$$
The second derivative of function $f(x)$ is:
$$f''(x) = e^x+1/x^2$$

```python
def ex2_derivative_2(x):
    return np.exp(x) + 1/(x**2)
```

```python
newton_method(2, ex2_function, ex2_derivative, ex2_derivative_2, x_ex2, y_ex2, initial_point_ex2, frame_speed = 5)
```
